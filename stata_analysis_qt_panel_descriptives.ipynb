```stata
egen max_collision_count = max(collision_count), by(intersection_id)
gen no_variation = ( max_collision_count == 0)
tab no_variation
tab collision_count if no_variation == 1

global list  collision_count personsinjured personskilled pedestriansinjured pedestrianskilled cyclistinjured cyclistkilled motoristinjured motoristkilled
foreach i of global list {
rename latenight_`i' late_`i'
}
global mean_list
global list  collision_count late_collision_count day_collision_count personsinjured personskilled pedestriansinjured pedestrianskilled cyclistinjured cyclistkilled motoristinjured motoristkilled late_personsinjured late_personskilled late_pedestriansinjured late_pedestrianskilled late_cyclistinjured late_cyclistkilled late_motoristinjured late_motoristkilled day_personsinjured day_personskilled day_pedestriansinjured day_pedestrianskilled day_cyclistinjured day_cyclistkilled day_motoristinjured day_motoristkilled
foreach i of global list {
	gen mean_`i' = `i'
	global mean_list $mean_list mean_`i'
}

```

```stata
collapse (sum)$list (mean) $mean_list year, by(flag_LPIS_ever quarterly)
```

```stata
twoway line collision_count quarterly if flag_LPIS_ever == 0, legend(lab(1 "Flag_LPIS_ever==0")) || line collision_count quarterly if flag_LPIS_ever == 1, legend(lab(2 "Flag_LPIS_ever==1")) ytitle("Sum of Collisions")
graph export "..\manuscripts\graph_of_time_trend.png", replace
```
